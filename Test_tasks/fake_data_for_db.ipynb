```python
from faker import Faker
```

```python
Faker.seed(0)
```

```python
fake = Faker()
```

```python
uids = list(fake.random_elements(elements=range(100,1000), length=100, unique=True))
```

```python
ages = list(fake.random_elements(elements=range(18,65), length=100))
```

```python
users_data = list(zip(uids, ages))
```

```python
users_data
```

```python
ituids = list(fake.random_elements(elements=range(1000,10000), length=250, unique=True))
```

```python
prices = list(fake.random_elements(elements=range(50,5100), length=250))
```

```python
items_data = list(zip(ituids, prices))
```

```python
items_data
```

```python
purchases_data = []
```

```python
for _ in range(500):
    purchases_data.append((*fake.random_elements(elements=range(10000,100000), length=1),
                  fake.random_element(elements=uids),
                  fake.random_element(elements=ituids),
                  str(fake.date_between(start_date='-5y'))))
```

```python
purchases_data
```
