# Task 2
Your task is to design a dashboard for a product manager. This dashboard should monitor user engagement and app usage. Which metrics should be included in the dashboard and why?

Before designing a dashboard, we should specify what we mean by engaged users. Let's assume that engaged users use the app regularly, are overall satisfied with the app, make transactions regularly and are willing to keep using it.
### Possible metrics for a dashboard:
* **DAU**, **MAU** - a successful app will be used by a large number of users and will attract more users over time
* **ASL**, **average number of sessions** - users will tend to spend more time in the app, if it's attractive
* **Retention rate** - users will tend to use the app more often, if it's successful
* **Conversion rate** - if the app's successful, CR will grow over time
* **Conversion window** - the window would be smaller if the app encourages users to make transactions sooner
* **ARPU**, **ARPPU** - if the app's successful, it will become more profitable
* **NPS (net promoter score)** - if the app's successful, more user will recommend it to others
* **Average review score on app stores** - if users are satisfied with the app, it will have a higher score. That could attract more new users as well
